```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("electricity_cost_dataset.csv")
df.head()
```

```python
df['electricity cost per resident'] = df['electricity cost'] / df['resident count']
df['electricity cost per resident'] = np.round(df['electricity cost per resident'], 3)
df.head()
```

```python
df['water consumption per resident'] = df['water consumption'] / df['resident count']
df['water consumption per resident'] = np.round(df['water consumption per resident'], 3)
```

```python
df['recycling efficency rate'] = df['recycling rate'] * df['utilisation rate']
df['recycling efficency rate'] = np.round(df['recycling efficency rate'], 3)
```

```python
df['resource intensity'] = (df['water consumption'] + df['electricty cost']) / df['residnet count']
df['resource intensity'] = np.round(df['resource intensity'], 3)
```

```python
df['resolution_efficiency'] = df['issue reolution time'] / df['resident count']
df['resolution_efficiency'] = np.round(df['resolution_efficiency'], 3)
```
