```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("electricity_cost_dataset.csv")
df.head()
```

```python
df['electricity cost per resident'] = df['electricity cost'] / df['resident count']
df['electricity cost per resident'] = np.round(df['electricity cost per resident'], 3)
```

```python
df['water_consumption_per_resident'] = df['water consumption'] / df['resident count']
df['water_consumption_per_resident'] = np.round(df['water_consumption_per_resident'], 3)
```

```python
df['recycling_efficency_rate'] = df['recycling rate'] * df['utilisation rate']
df['recycling_efficency_rate'] = np.round(df['recycling_efficency_rate'], 3)
```

```python
df['resource_intensity'] = (df['water consumption'] + df['electricity cost']) / df['resident count']
df['resource_intensity'] = np.round(df['resource_intensity'], 3)
```

```python
df['resolution_efficiency'] = df['issue reolution time'] / df['resident count']
df['resolution_efficiency'] = np.round(df['resolution_efficiency'], 3)
```

```python
df['air_quality_impact'] = df['air qality index'] * df['resident count']
df['air_quality_impact'] = np.round(df['air_quality_impact'], 3)
```

```python
df['resource_intensity'] = (df['water consumption'] + df['electricity cost']) / df['site area']
df['resource_intensity'] = np.round(df['resource_intensity'], 3)
```

```python
df['energy_efficiency'] = df['electricity cost'] / (df['site area'] * df['utilisation rate'])
df['energy_efficiency'] = np.round(df['energy_efficiency'], 3)
```

```python
df['water_electricity_ratio'] = df['water consumption'] / df['electricity cost']
df['water_electricity_ratio'] = np.round(df['water_electricity_ratio'], 3)
```

```python
df['population_density'] = df['resident count'] / df['site area']
df['population_density'] = np.round(df['population_density'], 3)
```

```python
df['issue_per_utility'] = df['issue reolution time'] / (df['water consumption'] + df['electricity cost'])
df['issue_per_utility'] = np.round(df['issue_per_utility'], 3)
```

```python
df['sustainability_score'] = (df['recycling rate'] + df['utilisation rate']) / df['air qality index']
df['sustainability_score'] = np.round(df['sustainability_score'], 3)
```

```python
df = pd.get_dummies(df, columns=['structure type'])
```

```python
df = df.astype({col: 'int' for col in df.select_dtypes('bool').columns})
```

```python
for col in df.columns:
    col_data = df[col]
    if np.issubdtype(col_data.dtype, np.number):
        # Replace +inf with max finite value
        if np.isposinf(col_data).any():
            finite_max = col_data[np.isfinite(col_data)].max()
            df[col].replace(np.inf, finite_max, inplace=True)
        # Replace -inf with min finite value
        if np.isneginf(col_data).any():
            finite_min = col_data[np.isfinite(col_data)].min()
            df[col].replace(-np.inf, finite_min, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.to_csv("Electricity Dataset.csv", index=False)
```
